# ***Prompt Engineering - "Asian Paints Concall" Case Study***

```python
#Updates the file path to GenAI folder in Google Drive
filepath = "/content/drive/MyDrive/GenAI/"   # Check the drive structure from 'Files' in google collab after mounting
```

```python
# open asian paints concall transcript saved as text file
with open(filepath + "AsianPaints.txt", "r") as f:
  transcript = ' '.join(f.readlines())

# print the transcript
print(transcript)
```

```python
# print length of transcript
print(len(transcript))

# Obtain length to understand how much cost it might incur based on number of tokens
```

## ***Question 1***

```python
base_instruction = '''You are a helpful assistant which helps financial analysts retrieve
relevant financial and business related information from documents.
Given below is a question and the transcript of an earnings call of an paints company, Asian paints,
which was attended by the top management of the firm.
Try to respond with specific numbers and facts wherever possible.
If you are not sure about the accuracy of the information, just respond that you do not know'''

# SYSTEM ROLE
# DESCRIPTION about the problem
# GIVING SPECIFIC INSTRUCTIONS
# TELL CHATGPT HOW TO BEHAVE

question = "How much asian paints business have grown?"

# I AM GIVING TRANSCRIPT BECAUSE CHATGOT DOES NOT HAVE LATEST INFO - BE COST AWARE AS INPUT TOKEN INCREASES
prompt = base_instruction + "\n\n" +  "Question: {0}".format(question) + "\n\n" + "Transcript: \n {0}".format(transcript)

print(prompt)
```

```python
# QUESTION 1
message = [{"role": "user", "content": prompt}]

chat_response = openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",
    messages=message,
    max_tokens=1000,
    temperature=0.5,
    n=1,
    stop=None,
    frequency_penalty=0,
    presence_penalty=0)

print(chat_response.choices[0].message.content)
```
